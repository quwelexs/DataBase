```python
excel_data = pd.ExcelFile("/content/[ПР №3] zfpo_rivne_region (3).xlsx")

# Читання окремих листів
group_entry_data = excel_data.parse('Вступ групи')
students_data = excel_data.parse('Здобувачі')

total_admissions = group_entry_data['Вступники'].sum()
print(f"Загальна кількість вступників: {total_admissions}")

if 'Заклад освіти' in group_entry_data.columns and 'Вступники' in group_entry_data.columns and \
   'Заклад освіти' in students_data.columns and 'Здобувачів' in students_data.columns:

    admissions_by_institution = group_entry_data.groupby('Заклад освіти')['Вступники'].sum()
    students_by_institution = students_data.groupby('Заклад освіти')['Здобувачів'].sum()

    combined_info = pd.DataFrame({
        'Вступники': admissions_by_institution,
        'Здобувачі': students_by_institution
    })

    combined_info['Відсоток'] = ((combined_info['Вступники'] / (combined_info['Вступники'] + combined_info['Здобувачі'])) * 100).round(2).astype(str) + '%'

    print(combined_info)
```
